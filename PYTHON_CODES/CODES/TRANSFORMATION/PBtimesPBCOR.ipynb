## Obtain the IMAGE of ALMA 

The ALMA's data, are corrected by Primary Beam Correction (PBcor). Para obtener la imagen obtenida Directamente de las antennas de ALMA, pero calibradas, se tiene que multiplicar por este Primary Beam (PB) y asi obtener la imagen resultante. 

$PB \times PBcor = IMAGE$

--------------------------- Import the package to use -----------------------------

```python
import numpy as np
from astropy.io import fits
```

```python
# Define the filenames of the two data cubes
#file1 = 'C:/Users/macka/OneDrive/Documentos/Master/PYTHON_CODES/RESULTS/FITS/MOMENT/CF_MOMENTUMS/PBcor_dividiendolos/moment0_CF+_N159_sci_VELAXIS_inK_SPECT_SPAT_RES_PB.fits'  # CUBE .pb
file1 = 'C:/Users/macka/OneDrive/Documentos/Master/PYTHON_CODES/FILES/CF+_DATAS/member.uid___A001_X3621_X1a6a.N159-13CII_sci.spw26.cube.I.pb.fits'   #CUBE .pb
file2 = 'C:/Users/macka/OneDrive/Documentos/Master/PYTHON_CODES/FILES/CF+_DATAS/member.uid___A001_X3621_X1a6a.N159-13CII_sci.spw26.cube.I.pbcor.fits'   #CUBE .pbcor
file3 = 'C:/Users/macka/OneDrive/Documentos/Master/PYTHON_CODES/RESULTS/FITS/MOMENT/CF_MOMENTUMS/NoPbCor/moment0_CF+_N159_sci_VELAXIS_inK_SPECT_SPAT_RES_3SIGMA_NOPBCor.fits'  # nopbcor


# Define the output filename
# output = 'C:/Users/macka/OneDrive/Documentos/Master/PYTHON_CODES/RESULTS/FITS/CF+_N159_NOPBcor.fits'
output = 'C:/Users/macka/OneDrive/Documentos/Master/PYTHON_CODES/RESULTS/FITS/CORRECT/CF+/nopbcor/member.uid___A001_X3621_X1a6a.N159-13CII_sci.spw26.cube.I.fits'
OVERWRITE = True  # Set to True to overwrite existing files


# Cargar los dos archivos FITS
with fits.open(file1) as hdul1:
    data1 = hdul1[0].data
    header1 = hdul1[0].header
    print(f"Cabecera del archivo PB:\n{repr(header1)}")
with fits.open(file2) as hdul2:
    data2 = hdul2[0].data
    header2 = hdul2[0].header
    print(f"Cabecera del archivo PBcor:\n{repr(header2)}")
with fits.open(file3) as hdul3:
    data3 = hdul3[0].data
    header3 = hdul3[0].header
    print(f"Cabecera del archivo PBcor:\n{repr(header3)}")
```

----------------------- Multiplication of cubes --------------------------------

```python
print(data1.shape)
print(data2.shape)
print(data3.shape)
```

### Transform FITS file with PB CORRECTED, to NO PBCorr.

```python

# Asegurarse de que las dimensiones de los datos sean iguales antes de multiplicar
if data1.shape != data2.shape:
    print("Las dimensiones de los datos no coinciden. No se puede multiplicar.")
else:
    # Multiplicar los datos
    # where data 2 correspond to PBco file, and data1 to PB file.
    resultado_multiplicacion = data2 * data1

    # Crear un nuevo archivo FITS para guardar el resultado
    hdu_nuevo = fits.PrimaryHDU(data=resultado_multiplicacion, header=header2)
    hdul_nuevo = fits.HDUList([hdu_nuevo])
    header_nuevo = hdu_nuevo.header

    print(repr(header_nuevo))
```

### Transform the image to PBcor.

```python
# Asegurarse de que las dimensiones de los datos sean iguales antes de multiplicar
if data1.shape != data3.shape:
    print("Las dimensiones de los datos no coinciden. No se puede multiplicar.")
else:
    # dividir los datos
    # where data 2 correspond to IMAGE (NO PBCOR) file, and data1 to PB file.
    resultado_division = data3 / data1

    # Crear un nuevo archivo FITS para guardar el resultado
    hdu_div = fits.PrimaryHDU(data=resultado_division, header=header3)
    hdul_div = fits.HDUList([hdu_div])
    header_div = hdu_div.header

    print(repr(header_div))
```

```python
# Crear el nuevo archivo FITS y copiar el encabezado de la primera imagen
hdu_final = fits.PrimaryHDU(data=resultado_multiplicacion, header=header2)
header_final = hdu_final.header
print(repr(header_final))
```
